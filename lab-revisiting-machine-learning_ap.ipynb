Check for **null values in all the columns**

Exclude the following variables by looking at the definitions. Create a new empty list called drop_list. We will append this list and then drop all the columns in this list later:

OSOURCE - symbol definitions not provided, too many categories
ZIP CODE - we are including state already
Identify columns that over 85% missing values

```python
#First drop the columns OSOURCE & ZIP CODE

data = data.drop(['OSOURCE', 'ZIP'], axis = 1)
```

```python
data.shape
```

```python
#now create an empty list called drop_list for later
drop_list = []
```

```python
#As we are working with categorical data lets gather it seperately
#We can already gather the categorical data
categorical = data.select_dtypes([object])
categorical.shape
```

Identify columns that over 85% missing values

Remove those columns from the dataframe

```python
#okay lets check for white spaces, and replace them with NaNs so that we can see for sure how many nans are there
categorical = categorical.apply(lambda x: x.replace(" ", np.NaN))#here is x is every value in df
categorical.head()
```

```python
#Then lets check how many columns are above 85 percent
nulls_percent_catdf = pd.DataFrame(categorical.isna().sum()/len(categorical)).reset_index()
nulls_percent_catdf.columns = ['column_name', 'nulls_percentage']
```

```python
#Here we see which columns have null percentage greater than 85%
nulls_percent_catdf[nulls_percent_catdf['nulls_percentage']>0.85]
```

```python
#add this to the drop_list 

drop_list = [ x for x in nulls_percent_catdf[nulls_percent_catdf['nulls_percentage']>0.85]['column_name']]
drop_list
```

```python
#drop those 30 columns
categorical = categorical.drop(drop_list, axis = 1)
```

```python
#Columns are now removed
categorical
```

Reduce the number of categories in the column GENDER. The column should only have either "M" for males, "F" for females, and "other" for all the rest

Note that there are a few null values in the column. We will first replace those null values using the code below:

print(categorical['GENDER'].value_counts())

```python
categorical['GENDER'].isna().sum()
```

```python
#First dealing with the NaNs in the Gender column as explained in lab instructions
categorical['GENDER'] = categorical['GENDER'].fillna('F')
```

```python
print(categorical['GENDER'].value_counts())
```

```python
#Now grouping all others together
categorical['GENDER'] = np.where(categorical['GENDER'].isin(['J','C','A', 'U']) , 'other', categorical['GENDER'])
```
